```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
path = '/Users/lailaa/Documents/GitHub/graphing-assignment-lailaaholmes'
murder = pd.read_csv('murderpedia_dataset.csv')
```

```python
murder
```

```python
#Graph 1 will be a line one and show the methods and count 
```

```python
method_count = ( murder[['method', 'gender']] .groupby('method') .count().reset_index().rename(columns={'gender': 'count'}))
```

```python
method_count
```

```python
plt.plot(method_count['method'], method_count['count'])  # The plot

plt.xlabel('Method of Killing')  # x label
plt.ylabel('Number of Offenders')  # y label

plt.title("Killers by Method")  # title

plt.xticks(rotation=45)  # rotates labels so they don’t overlap

plt.show()
```

```python
#Second one will be a bar graph with gender and method
```

```python
gender_method = (murder.groupby(['method', 'gender']).size() .reset_index(name='count'))
pivot_df = gender_method.pivot(index='method', columns='gender', values='count').fillna(0) 
#https://www.geeksforgeeks.org/python/python-pandas-pivot (used this for the pivot info)
```

```python
gender_method
```

```python
male_bar = pivot_df["Male"]
female_bar = pivot_df["Female"]
```

```python
pivot_df = gender_method.pivot(index='method', columns='gender', values='count').fillna(0)
pivot_df
```

```python
import numpy as np

barWidth = 0.25

r1 = np.arange(len(male_bar))
r2 = [x + barWidth for x in r1]

# Make the plot
plt.bar(r1, male_bar, color='tab:blue', width=barWidth, edgecolor='white', label='Male')
plt.bar(r2, female_bar, color='tab:pink', width=barWidth, edgecolor='white', label='Female')

# X labels (methods)
plt.xlabel('Method of Killing', fontweight='bold')
plt.xticks([r + barWidth/2 for r in range(len(male_bar))], pivot_df.index, rotation=45)

# Y + title
plt.ylabel('Number of Offenders')
plt.title("Killing Methods by Gender")

# Legend
plt.legend()

plt.show()

#https://stackoverflow.com/questions/61183826/how-to-increase-the-bar-width-in-a-grouped-bar-plot-without-reducing-the-gap-bet#:~:text=%23%20libraries%20import%20numpy%20as%20np,legend()%20plt.show() 
#Used this info 
```

```python
#Graph 1
#This graph shows how often each method of killing appears in the dataset. 
#You can see that more violent methods, especially shooting, happen a lot more than others, so some methods are clearly more common.
```

```python
#Graph 2
#This graph compares how males and females differ across methods of killing. 
#It shows that males use more violent methods like shooting way more, while females are more spread out across methods, especially poisoning.
```

```python
#Next doing the graphs side by side 
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(14,6))
# -------- Graph 1 --------
plt.subplot(1, 2, 1)

plt.bar(method_count ['method'], method_count ['count'])

plt.xlabel('Method of Killing')
plt.ylabel('Number of Offenders')
plt.title('Killers by Method')

plt.xticks(rotation=45)


# -------- Graph 2 --------
plt.subplot(1, 2, 2)

barWidth = 0.3
r1 = np.arange(len(male_bar))
r2 = [x + barWidth for x in r1]

plt.bar(r1, male_bar, width=barWidth, label='Male')
plt.bar(r2, female_bar, width=barWidth, label='Female')

plt.xlabel('Method of Killing')
plt.ylabel('Number of Offenders')
plt.title('Killing Methods by Gender')

plt.xticks([r + barWidth/2 for r in range(len(male_bar))], pivot_df.index, rotation=45)

plt.legend()

# spacing so labels don’t overlap
plt.tight_layout()

plt.show()

#https://www.geeksforgeeks.org/python/place-plots-side-by-side-in-matplotlib/ (used this to get the code)
```

```python
#My last graph will be a piechart 
```

```python
violence_pie = (murder.groupby('violent_vs_nonviolent') .size().reset_index(name='count'))

tot = violence_pie['count'].sum()
violence_pie['% total'] = violence_pie['count'] / tot

violence_pie
```

```python
fig, ax1 = plt.subplots()


colors = ['#ff9999','#66b3ff','#99ff99']  # pink, blue, green

explode = (0.1, 0, 0)  # highlight first category

ax1.pie( violence_pie['% total'],labels=violence_pie['violent_vs_nonviolent'], autopct='%1.1f%%',explode=explode, colors=colors,shadow=False,startangle=90,pctdistance=0.85)


plt.title("Violent vs Non-Violent Methods (%)", fontsize=14)


ax1.legend(labels=violence_pie['violent_vs_nonviolent'],loc='upper left',bbox_to_anchor=(1, 1))

plt.show()
```
